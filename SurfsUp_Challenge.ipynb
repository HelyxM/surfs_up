```python
# Dependencies
import numpy as np
import pandas as pd
import datetime as dt

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
results = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
```

```python
#  3. Convert the June temperatures to a list.
june_results = list(results)
# june_results
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_results, columns=['date', 'June Temps'])
june_df.set_index(june_df['date'], inplace=True)
june_temps = june_df.drop(columns=['date'])
june_temps.head()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temps.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
results_2 = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()
```

```python
# 7. Convert the December temperatures to a list.
december_results = list(results_2)
# december_results
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_df = pd.DataFrame(december_results, columns=['date', 'December Temps'])
december_df.set_index(december_df['date'], inplace=True)
december_temps = december_df.drop(columns=['date'])
december_temps.head()
```

```python
# 9. Calculate and print out the summary statistics for the December temperature DataFrame.
december_temps.describe()
```

```python
june_results = session.query(Measurement.date, Measurement.prcp, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
june_results_list = list(june_results)
june_weather_df = pd.DataFrame(june_results_list, columns=["date", "precipitation", "temperature"])
june_weather_df.describe()
```

```python
december_results = session.query(Measurement.date, Measurement.prcp, Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()
december_results_list = list(december_results)
december_weather_df = pd.DataFrame(december_results_list, columns=["date", "precipitation", "temperature"])
december_weather_df.describe()
```

```python
december_years = december_weather_df["date"].str.split('-', expand=True)
december_weather_df["date"] = december_years[0]
december_years_df = december_weather_df.loc[:, ["date", "precipitation", "temperature"]]
december_averages = december_years_df.groupby(december_years_df["date"]).mean()
december_averages
```
